## EXCEL

we have a file called  `movies_db.xlsx `, lets read this file and do some operations with it

Install an extention to read excel files `pip install openpyxl`

read the file and store it in a dataframe

```python
df_movies_db = pd.read_excel('movies_db.xlsx') # reads an Excel file named 'movies_db.xlsx' into a new DataFrame called df_movies_db

df_movies_db.head() # displays the first 5 rows of the df_movies_db DataFrame
```

if we see the excel file we can see that it has multiple sheets, 

```python
df_movies_db_sheet2 = pd.read_excel('movies_db.xlsx', sheet_name='financials') # reads the 'Sheet2' from the 'movies_db.xlsx' Excel file into a new DataFrame called df_movies_db_sheet2
df_movies_db_sheet2.head() # displays the first 5 rows of the df_movies_db_sheet2 DataFrame
```

In the currency col, we can see USD, $$, Dollars. lets make all of them to USD

```python
def std_currency(x):
    if x == "Dollars" or x == "$$":
        return "USD"
    else:
        return x

df_movies_db_financials = pd.read_excel('movies_db.xlsx', sheet_name='financials', converters={
    'currency': std_currency
})

df_movies_db_financials.head()
```

we have fixed a issue in our sheet, now lets merge the financial sheet with movies sheet, we know the movie_id 

```python
df_merged_moviesAndFinancials = pd.merge(df_movies_db, df_movies_db_financials, on='movie_id') # merges the df_movies_db and df_movies_db_financials DataFrames based on the 'movie_id' column and creates a new DataFrame called df_merged_moviesAndFinancials
df_merged_moviesAndFinancials.head() 
```

We can make a new excel file with this new data frame


```python
df_merged_moviesAndFinancials.to_excel('merged_movies_and_financials.xlsx', sheet_name='merged_data', index=False)

# saves the df_merged_moviesAndFinancials DataFrame to a new Excel file named 'merged_movies_and_financials.xlsx' without including the index column
```

## Handling Missing Data

We data we find in real life is not always complete there may be many fields that contains trash, empty or random values, if we do not treat(handle) these values they will effect our final results.

The file `weather_data.csv` contains such values, lets see how we van handle them

```python
df = pd.read_csv("weather_data.csv")
df.head()
```

We can see multiple fields containing `NaN` values, in real world a user might input random values like 999 or 69 in temprature field which will have a big impact on our analysis, like our avg(temprature) value will increase drastically. 

```python
df["day"].dtype

```

- type convert

    The values of day cols are of string type, we need them to be of  datatype - date

```python
df["day"] = pd.to_datetime(df["day"])
```

```python
type(df.temperature[0])
```

- Fill NaN values

```python
df_eg = df.fillna('0') # Replaces NaN values with 0
df_eg
```

Ideally we should be filling NaN values with something meaningfull, like avg, mean, min, max, etc to make sure we can better overall results from our analysis  

```python
df.fillna({
    'temperature' : df.temperature.mean(),
    'windspeed' : df.windspeed.mean(),
    'event' : "Regular"
})
```

One other approach is to fill NaN values with the values of the previous field - fill forward

```python
df.ffill()
```

Backward fill

```python
df.bfill()
```

```python
df.bfill(axis=1)

# Fills sideways i.e along cols 
```

- Interepolation

`interpolate()` method can be used to fill missing (NaN) numerical data in a Series or DataFrame by estimating values between known data points. 

***Does not work on str***

```python
df.select_dtypes(include=['number', 'datetime']).interpolate()

# By default the interpolate() assumes a linear relationship, we can chagne this method.
```

- **Delete/Drop** the rows with NaN Values

If the rows with NaN values are insignificant, we can drop and to avoid the hassle

```python
df.dropna()
```

```python
df.select_dtypes(exclude=["datetime"]).dropna(how="all")
# Drops rows with where all fields are NaN
```

```python
df.dropna(thresh=2)
# only keeps rows with atleat 2 filds with not nan values.

# We can chagne axis to drop cols
```

## Grouping Data

Consider the data set below

```python
df = pd.read_csv("weather_by_cities.csv")
df
```

We can see in cities column, we have same cities appearing multiple times.

Let's say we have to find min/max temprature in a particular city or we have to do some analysis to a particular city.

We can use `groupby()` method to group data by one or more **keys**(like cities)

```python
g = df.groupby("city")
g
```

return a groupby object which is iterable

```python
for city, data in g:
    print(city)
    print(data)
```

As we can see, in our object `g`, in each iteration which have 2 parts:
1. Key - `city` 
2. dataframe-`data`.

we can furher unpack the df `data`  to access specific cols and fields

```python
for city, data in g:
    print(city)
    print(data.temperature.max())
```

- get group

```python
g.get_group("mumbai")
```

- Apply common methods to a group

```python
g.max()
```

```python
g.min()
```

```python
g.describe()
```

```python
g.size()
```

### Custom Grouping

lets say you we want to group our data based on following tempratures:
 1. temp between 80 and 90
 2. temp between 50 adn 60
 3. other temp

```python
import numpy as np

conditions = [
    ((df["temperature"] > 79) & (df["temperature"] < 91)),
    ((df["temperature"] > 49 ) & (df["temperature"] < 61)),
]

names = ["80-90", "50-60"]

g = df.groupby(np.select(conditions, names, default="Others"))
```

```python
for key, data in g:
    print(key)
    print(data)
    print()
```

## Concatenation and Merging

Consider the below 2 separate dfs 

```python
india_weather = pd.DataFrame({
    "city": ["mumbai","delhi","banglore"],
    "temperature": [32,45,30],
    "humidity": [80, 60, 78]
})
india_weather
```

```python
us_weather = pd.DataFrame({
    "city": ["new york","chicago","orlando"],
    "temperature": [21,14,35],
    "humidity": [68, 65, 75]
})
us_weather
```

Now lets concatenate them into a single df

- concatenate

```python
df = pd.concat([india_weather, us_weather], ignore_index=True)
df
```

- Merge

Consider we have 2 dfs with same cities but different Cols, like below 

```python
df1 = pd.DataFrame({
    "city": ["new york","chicago","orlando"],
    "temperature": [21,14,35],
})
df1
```

```python
df2 = pd.DataFrame({
    "city": ["chicago","new york","orlando"],
    "humidity": [65,68,75],
})
df2
```

```python
df = pd.merge(df1, df2, on="city")
df
```

By default the merge methods performs an `inner join` - read more about them including - left, right and full join
